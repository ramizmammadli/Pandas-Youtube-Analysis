```python
#Finding the ratio of likes to overall reactions and sorting it in descending order.
```

```python
# method no:1
```

```python
# function for finding ratio.
def ratio(vec):
    l = vec[0]
    dl = vec[1]
    if (l+dl != 0):
        return float(l/(l+dl))
    return 0.0
```

```python
dataset["likes_dislikes_ratio_method1"] = dataset[["likes", "dislikes"]].apply(ratio, axis = 1)
# new column is created and the ratio function is applied to the sub-dataset using vector. (Faster, more memory-friendly but not quite detailed.)
```

```python
dataset.sort_values("likes_dislikes_ratio_method1", ascending = False, inplace = True) # then sorted in descending order.
```

```python
def likesdislikesratio(likes, dislikes):
    likesList = []
    # 'dataset["likes"].iteritems()' is a zip file of indices and the number of likes for each video. 
    # it is converted to list in order to be iterated. 
    # Value points to the number of likes, and key is index, which is not useful for us. 
    # Briefly, the values (the number of likes) are iteratively put in a list.
    for key, value in list(dataset["likes"].iteritems()):  
        likesList.append(value)
    # same logic happens here for dislikes.
    dislikesList = []
    for key, value in list(dataset["dislikes"].iteritems()):
        dislikesList.append(value)
    
    # the lists of likes and dislikes are zipped together.
    likedislike = list(zip(likesList, dislikesList))
    ratio = []
    # then the wanted ratio will be found below.
    for like,dislike in likedislike:
        if (like+ dislike) == 0: # for avoiding division to zero
            ratio.append(0)
        else:
            ratio.append(like / (like+dislike))
    return ratio
            
dataset["likes_dislikes_ratio_method2"] = likesdislikesratio(dataset["likes"], dataset["dislikes"]) # added to the dataset
# (slower, less memory-friendly, but quite detailed to understand.)
dataset.sort_values("likes_dislikes_ratio_method2", ascending = False, inplace = True) # sorted.
```
